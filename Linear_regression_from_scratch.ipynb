Linear Regression From Scratch:

```python
# Describing class:
class reg:
    def __init__(self):
        self.m=None
        self.b=None
        
    def fit(self,x_train,y_train):
        num=0
        den=0
        mean_x=x_train.mean()
        mean_y=y_train.mean()
        for i in range (x_train.shape[0]):
            x=x_train[i] - mean_x
            y=y_train[i] - mean_y
            num=num+(x*y)  
            den=den+(x**2)
        self.m=num/den
        self.b=mean_y - ((self.m) * mean_x)          
        
    def pred(self,x_test):
        return self.m*x_test + self.b
```

```python
#importing Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('placement.csv')
df.head()
```

```python
X=df.iloc[:,0].values
Y=df.iloc[:,1].values
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
lr=reg()
lr.fit(x_train,y_train)
```

```python
my_pred=lr.pred(x_test[0])
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train.reshape(-1,1), y_train)

print("My Slope:", lr.m)
print("Sklearn Slope:", model.coef_)
print("My intercept:", lr.b)
print("Sklearn intercept:", model.intercept_)
```

```python
model_pred=model.predict(x_test[0].reshape(-1,1))
print('sklearn predicted:',model_pred)
print('my predicted value:',my_pred)
```
